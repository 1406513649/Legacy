## 5. Object Oriented FEM
Our `ezfem` code, up until now, would be considered "procedural", meaning our code marches through the FEM algorithm step by step using functions, or "procedures", to complete a given task.  Using several small special purpose functions, our code is also highly modularized.  In contrast, "object oriented" programing carries out the same task by using objects.  Objects have "methods" (associated functions) and "attributes" (associated variables or constants).  Object oriented programming allows for great flexibility in adding new components to a program or building on top of existing components.  A finite element program is an ideal candidate for object oriented design.

In this exercise, we will implement an object oriented design to our `ezfem` code by creating and using an element class.  Specific elements (linear, quadratic, beam, etc.) will be implemented as sub-classes of a base class `Element`.  Our code will then use `Element` objects to carry out the element related computations.

It is also common to partition a mesh in to element blocks (SOLID SECTIONS in Abaqus).  Each element block is made up of a collection of like elements.  Our code will be modified to treat the finite element mesh as a collection of element blocks.

```python
def gen_elem_blks(*args, **kwargs):
    """Generate element blocks

    Parameters
    ----------
    args : tuple of lists
        args[i] is (length, number of elements, element name,
                    element area, element properties)
        for the ith block.  The element name is sent to get_elem_by_name
        to determine the element type.
    kwargs : dict
        Optional keywords.
        offset : real
            Coordinates shall be offset by this much

    Returns
    -------
    elem_blks : list of lists
        elem_blks[i] is (element block number, elements in block,
                         element type, element area, element properties)

    Example
    -------
    >>> elem_blk_1 = (4., 2, "Link2", a1, prop)
    >>> elem_blk_2 = (3., 3, "Link2", a2, prop)
    >>> coords, conn, elem_blocks = gen_elem_blocks(elem_blk_1, elem_blk_2)

    """
    xa = 0.
    coords = np.array(xa)
    conn = []
    elem_blks = []
    num_elem = 0
    num_node = 0
    for (i, item) in enumerate(args):
        try:
            # l : length of element block
            (len_elem_blk, num_elem_in_blk, elem_blk_elem,
             elem_blk_area, elem_blk_prop) = item
        except ValueError:
            raise SystemExit("expected 5 items per arg, got {0}".format(len(item)))

        # get coords and connectivity for this block
        xb = xa + len_elem_blk
        elem_blk_coords, elem_blk_conn = linmesh_1d(xa, xb, num_elem_in_blk)
        elem_blk_conn += num_elem

        # add this blocks coords and connectivity to global coords and
        # connectivity
        coords = np.append(coords, elem_blk_coords[1:])
        conn.extend(elem_blk_conn + num_node)

        # set up this element block
        elem_type = get_elem_by_name(elem_blk_elem)
        elems_in_blk = np.arange(len(elem_blk_conn)) + num_elem
        elem_blk = [i, elems_in_blk, elem_type, elem_blk_area, elem_blk_prop]
        elem_blks.append(elem_blk)

        xa = xb
        num_elem += len(elems_in_blk)
        num_node += len(elem_blk_coords)

    # fix up the connectivity, putting -1 in empty node locations that arise,
    # for instance, for a mesh containing both 3 node quadratic elements and 2
    # node linear
    max_num_node = max(len(a) for a in conn)
    for (i, a) in enumerate(conn):
        conn[i] = [a[j] if j < len(a) else -1 for j in range(max_num_node)]
    conn = np.array(conn, dtype=np.int32)

    offset = kwargs.get("offset", 0.)

    return coords + offset, conn, elem_blks
```
